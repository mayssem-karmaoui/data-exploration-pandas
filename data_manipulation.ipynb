#II. Extraction de donnÃ©es d'une BD

```python
# 1
import sqlite3
import pandas as pd
# 2. crÃ©er une base de donnÃ©e
db_path = '/content/drive/MyDrive/tp3DS/testTP.db'
conn = sqlite3.connect(db_path)
```

```python
# 3. crÃ©er  table table "users"
cursor = conn.cursor()
cursor.execute('''
    CREATE TABLE IF NOT EXISTS users (
        id INTEGER PRIMARY KEY,
        nom TEXT NOT NULL,
        age INTEGER NOT NULL
    )
''')
conn.commit()  # Sauvegarder
```

```python
# 4. Inserer data
cursor.execute("INSERT INTO users (nom, age) VALUES ('Alice', 30)")
cursor.execute("INSERT INTO users (nom, age) VALUES ('Bob', 25)")
cursor.execute("INSERT INTO users (nom, age) VALUES ('Charlie', 35)")
conn.commit()
```

```python
# 5. Inserer data avec dictionaire
user_data = {'nom': 'David', 'age': 28}
cursor.execute("INSERT INTO users (nom, age) VALUES (:nom, :age)", user_data)
conn.commit()
```

```python
#6. Afficher les donnÃ©es ainsi ajoutÃ©es
cursor.execute("SELECT * FROM users")
all_users = cursor.fetchall()
print("Data in 'users' table:")
for user in all_users:
    print(user)
```

```python
# 7.  rÃ©cupÃ©rer le contenu de la table crÃ©e et afficher le ce dataframe.
df_users = pd.read_sql_query("SELECT * FROM users", conn)
print("\nDataFrame from 'users' table:\n", df_users)
```

```python
import requests
import urllib.request
import time
import csv
from bs4 import BeautifulSoup
import pandas as pd
url='http://books.toscrape.com/'
reponse=requests.get(url)
print(response)
```

```python
page_contents= response.text
print(page_contents)
```

```python
doc = BeautifulSoup(page_contents,"html.parser")
print(doc)
```

```python
book_title_tags = doc.find_all('h3')
book_titles =[]
for tags in book_title_tags:
  book_titles.append(tags.text)
print(book_titles)
```

```python
book_price_tags = doc.find_all('p',class_='price_color')
book_price = []

for tags in book_price_tags:
  book_price.append(tags.text.replace('Ã',''))
print(book_price)
```

```python
book_stock_tags = doc.find_all('p',class_='instock availabilily')

book_stock=[]
for tags in book_stock_tags:
  book_stock.append(tags.text.strip())
print(book_stock)
```

```python
titles = doc.find_all('h3')
book_url=[]
for article in titles:
  for link in article.find_all('a',href=True):
    url= link['href']
    full_link = 'https://books.toscrape.com/' + url
    if full_link not in book_url:
      book_url.append(full_link)
print(book_url)
```

```python
book_data = pd.DataFrame({
   "Title":  book_titles,
   "Price": book_price,
   "Availability" : book_stock,
   "URL" : book_url
})

print(book_data)
```
